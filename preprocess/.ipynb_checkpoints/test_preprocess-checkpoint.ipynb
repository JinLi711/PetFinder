```python
train_dir = "../data/train_images/"
imgs = ["0a0e8c15b-1.jpg", "0a0e8c15b-2.jpg",
        "00a1a8af4-1.jpg", "0a9f9f178-8.jpg"]
# np.array (Image.open(train_dir + "0a0e8c15b-1.jpg"))


def create_np_from_imgs(img_names, img_dir, resize_shape=(32, 32)):
    """
    Open images from a directory. 
    Combine them together into numpy array.

    :param img_names: list of image file names
    :type  img_names: list
    :param img_dir: path of directory leading to images
    :type  img_dir: str
    :param resize_shape: dimensions to reshape image
    :type  resize_shape: tuple
    :returns: array of the images together.
              Shape: (len(img_names), resize_shape, 3)
    :rtype:   array
    """

    from PIL import Image
    import cv2
    np_imgs = []
    for img_name in img_names:
        image = np.array(Image.open(img_dir + img_name))
        resized_image = cv2.resize(image, resize_shape)
        np_imgs.append(resized_image)
    np_imgs = np.array(np_imgs)
    return np_imgs


# test = np.array([np.array(Image.open(train_dir + file)) for file in imgs ])
# test.shape
test = create_np_from_imgs(imgs, train_dir)
```

```python
def triple_generator(
    generator, 
    tabular, 
    text, 
    images, 
    col_name,
    label_name,
    path_dir,
    batch_size=10,
    ):
    """
    Generator for gathering batches for tabular data, 
    text data, and image data.
    
    :param generator: image generator from keras
    :type  generator: keras_preprocessing.image.ImageDataGenerator
    :param tabular: an array of the tabular data
    :type  tabular: numpy.ndarray
    :param text:
    :type  text:
    :param images: dataframe where one column contains 
                   the image file names and another column
                   contains the labels
    :type  images: pandas.core.frame.DataFrame
    :param col_name: name of images column name
    :type  col_name: str
    :param label_name: name of label column
    :type  label_name: str
    :param path_dir: directory path of images
    :type  path_dir: str
    :param batch_size: size of the batch size
    :type  batch_size: int
    :yields: 
    :ytype:  (list, list)
    """
    
    while True:
        idx = np.random.permutation( images.shape[0])
        
        batches = generator.flow_from_dataframe(
            dataframe=images.loc[idx,],
            directory=path_dir,
            target_size=(32, 32),
            x_col=col_name,
            y_col=label_name,
            batch_size=batch_size,
            seed=7
        )
        
        idx0 = 0
        for batch in batches:
            idx1 = idx0 + batch[0].shape[0]

            yield [ tabular[ idx[ idx0:idx1 ] ], batch[0]], batch[1]

            idx0 = idx1
            if idx1 >= images.shape[0]:
                break
#     gen_tabular = generator.flow(
#         tabular,
#         label,
#         batch_size=batch_size,
#         seed=7
#     )
    
#     datagen = ImageDataGenerator(rescale=1./255.)
#     while True:
#         gen_image = generator.flow_from_dataframe(
#             dataframe=images,
#             directory="../data/train_images/",
#             target_size=(32, 32),
#             x_col='Img_Name',
#             y_col='AdoptionSpeed',
#             batch_size=batch_size,
#             seed=7
#         )
    
#     while True:
#         Yi1 = gen_tabular.next()
#         Yi2 = gen_image.next()
#         yield {'tabular': Yi1[0], 'images': Yi2[0]}, Yi1[1]
#         yield Yi2[0], Yi2[1]
#         yield Yi1[0], {'class': Yi1[1], 'superclass': Yi2[1]}
```

```python
from keras_preprocessing.image import ImageDataGenerator
train_datagen = ImageDataGenerator(
    rescale=1./255.
)

from keras import layers
from keras import Input
from keras.models import Model

def create_model():
    tabular = Input(
        shape=(85,),
        name="tabular"
    )
    
    tab = layers.Dense(
        4,
        activation="relu"
    )(tabular)
    
    img = Input(shape=(32,32,3))
    
    x = layers.Conv2D(
        32, 
        (6,6),
        activation='relu'
    )(img)
    
    x = layers.MaxPool2D(
        (4,4),
        padding="same"
    )(x)
    
    x = layers.Flatten(
    )(x)
    
    x = layers.Dense(
        4,
        activation="relu"
    )(x)
    
    result = layers.concatenate(
        [tab, x],
        axis=-1
    )
    result = layers.Dense(
        5, 
        activation="softmax"
    )(result)
    
    model = Model(
        [tabular, img],
        result
    )
#     model = Model(
#         img,
#         x
#     )
    return model
```

```python
model = create_model()
```

```python
# model.summary()
```

```python
from keras import optimizers
model.compile(
    optimizer=optimizers.RMSprop(lr=1e-4),
    loss='categorical_crossentropy',
    metrics=['accuracy']
)
```
